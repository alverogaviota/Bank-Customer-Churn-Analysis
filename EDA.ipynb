```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/customer.csv')
```

```python
display(df.isnull().sum())
display(df.duplicated().sum())
```

```python
fig, ax = plt.subplots()

label = df['Exited'].unique()

ax.pie(df['Exited'].value_counts(sort=False), autopct='%1.2f%%', startangle=-30, labels=label, colors=('crimson', 'dodgerblue'))
ax.set_title("Percentage of customer's churn")
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.countplot(df, x='Geography', hue='Exited', palette='dark:blue_r')
ax.set_title("Customer's Churn Based On Geography")
ax.grid(axis='y', linestyle='--', alpha=0.2)

plt.show()
```

```python
df.groupby(['Geography', 'Exited']).size().unstack()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.countplot(df, x='Gender', hue='Exited', palette='dark:indigo_r')
ax.set_title("Customer's Churn Based On Gender")
ax.grid(axis='y', linestyle='--', alpha=0.2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.histplot(df, x='Age', hue='Exited', multiple='stack', palette='dark:teal_r')
ax.set_title("Customer's Age Distribution")

plt.show()
```
